```python
import tensorflow as tf
import numpy as np
```

# Load image 

```python
from tensorflow.keras.preprocessing import image
img=image.load_img("E:\\bugtti.jpg",target_size=(224,224))
img
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.imshow(img)
```

# Preprocessing

```python
resized_img=image.img_to_array(img)
final_image=np.expand_dims(resized_img,axis=0)  # need fourth dimension
final_image=tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
resized_img.shape
```

```python
final_image.shape
```

# MobileNet_v1

```python
mobile=tf.keras.applications.mobilenet.MobileNet()
```

```python
prediction=mobile.predict(final_image)
```

```python
print(prediction)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results=imagenet_utils.decode_predictions(prediction)
```

```python
print(results)
```

# MobileNet_v2

```python
mobile2=tf.keras.applications.mobilenet_v2.MobileNetV2()
```

```python
prediction2=mobile2.predict(final_image)
```

```python
results2=imagenet_utils.decode_predictions(prediction2)
```

```python
print(results2)
```
